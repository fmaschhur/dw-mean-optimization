```python
import pandas as pd
```

```python
stats_file = "/Users/Max/Dropbox/Uni/Master/SSPJ/ssp-dtw-mean-optimization/reference/DataSummary.csv"
```

```python
df_stats = pd.read_csv(stats_file)
```

```python
df_stats.columns
```

```python
df_stats['Samples'] = df_stats['Train '] + df_stats['Test ']

columns = ['Name', 'Length', 'Samples', 'Class']
df_stats.drop(columns=[c for c in list(df_stats.columns) if c not in(columns)], inplace=True)
df_stats = df_stats[columns]
df_stats
```

```python
df_stats = df_stats[df_stats.Name.isin(["Adiac", "Beef", "CBF", "ChlorineConcentration", "Coffee", "ECG200", "ECG5000", "ElectricDevices", "FaceAll", "FaceFour", "FiftyWords", "Fish", "GunPoint", "Lightning2", "Lightning7", "OSULeaf", "OliveOil", "PhalangesOutlinesCorrect", "SwedishLeaf", "SyntheticControl", "Trace", "TwoPatterns", "Wafer", "Yoga"])].set_index('Name')
df_stats
```

```python
print(df_stats.to_latex())
```

```python
df_stats['Length'] = pd.to_numeric(df_stats['Length'])
df_stats['Samples'] = pd.to_numeric(df_stats['Samples'])
df_stats['Class'] = pd.to_numeric(df_stats['Class'])
```

```python
df_stats_meta = df_stats.agg(['mean', 'median', 'min', 'max'])
print(df_stats_meta.to_latex(float_format="{:0.2f}".format))
#df_stats_meta
```
